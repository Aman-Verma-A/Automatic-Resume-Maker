```python
from fpdf import FPDF

class PDF(FPDF):
    def header(self):
        # Register the regular and bold versions of DejaVuSans
        self.add_font('DejaVu', '', 'C:/Users/LENOVO/fonts/DejaVuSans.ttf', uni=True)
        self.add_font('DejaVu', 'B', 'C:/Users/LENOVO/fonts/DejaVuSans-Bold.ttf', uni=True)
        self.set_font('DejaVu', '', 16)
        self.cell(0, 10, 'Resume', border=False, ln=1, align='C')

    def section_title(self, title):
        # Use the bold font for section titles
        self.set_font('DejaVu', 'B', 12)
        self.set_text_color(0, 0, 255)  # Blue color for headings
        self.cell(0, 10, title, ln=1, align='L')
        self.ln(5)

    def add_text(self, text):
        # Use the regular font for body text
        self.set_font('DejaVu', '', 12)
        self.set_text_color(0, 0, 0)
        self.multi_cell(0, 10, text)
        self.ln(5)

def create_resume():
    pdf = PDF()
    pdf.add_page()

    # Collect User Details
    name = input("Enter your full name: ")
    email = input("Enter your email: ")
    contact = input("Enter your contact number: ")
    linkedin = input("Enter your LinkedIn profile URL: ")
    github = input("Enter your GitHub profile URL: ")
    portfolio = input("Enter your portfolio URL (if any): ")

    # Add Contact Info
    pdf.set_font('DejaVu', 'B', 12)
    pdf.cell(0, 10, name, ln=1)
    pdf.set_font('DejaVu', '', 12)
    pdf.cell(0, 10, f"Email: {email} | Phone: {contact}", ln=1)
    pdf.cell(0, 10, f"LinkedIn: {linkedin} | GitHub: {github} | Portfolio: {portfolio}", ln=1)
    pdf.ln(10)

    # Objective
    pdf.section_title("Objective")
    objective = input("Enter your career objective: ")
    pdf.add_text(objective)

    # Education
    pdf.section_title("Education")
    course = input("Enter your recent course (e.g., B.tech , BCA etc): ")
    university = input("Enter your university name: ")
    grad_year = input("Enter your graduation year: ")
    universityad = input("Enter your university Address: ")
    schooling = input("Enter your schooling details (e.g., High School, XYZ School, 2015): ")
    pdf.add_text(f"{course}, {university} (Graduated: {grad_year} ,{universityad})")
    pdf.add_text(schooling)

    # Skills
    pdf.section_title("Skills")
    skills = input("Enter your skills (comma-separated): ")
    pdf.add_text(skills)

    # Experience
    pdf.section_title("Experience")
    experience = input("Describe your Work Experience with Company name and Year of Experience: ")
    pdf.add_text(experience)

    # Internships
    pdf.section_title("Internships")
    internship = input("Describe your Internships with Starting and Ending Date and Name of Company: ")
    pdf.add_text(internship)

    # Projects
    pdf.section_title("Projects")
    projects = input("Describe your projects In Summary (use commas to separate multiple projects): ")
    pdf.add_text(projects)

    # Certifications
    pdf.section_title("Certifications")
    certifications = input("List your certifications: ")
    pdf.add_text(certifications)

    # Publications
    pdf.section_title("Publications")
    publications = input("List your publications: ")
    pdf.add_text(publications)

    # References
    pdf.section_title("References")
    pdf.add_text("Available upon request.")

    # Save PDF
    output_file = f"{name.replace(' ', '_')}_Resume.pdf"
    pdf.output(output_file)
    print(f"Resume successfully created! Saved as {output_file}")

# Run the Resume Creator
if __name__ == "__main__":
    create_resume()
```
